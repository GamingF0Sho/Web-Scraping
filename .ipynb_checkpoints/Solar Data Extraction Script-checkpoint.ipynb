```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
clist = pd.read_csv("solar.csv")
print(clist.head())
```

```python
cname = clist["Company Name"]
cname_list = []
for c in cname:
    string = c.replace(' ','_')
    url_st = ''.join(["http://en.openei.org/wiki/", string])
    cname_list.append(url_st)
test_list = (cname_list[0:5])
print(test_list)
```

```python
scrape_list = []
scrape_data = []
scrape_list_test = []
for url in cname_list:
    temp_list = []
    temp_data = []
    response = requests.get(url)
    content = response.content
    parser = BeautifulSoup(content, 'html.parser')
    body = parser.find_all("body")
    th = body[0].find_all("th")
    td = body[0].find_all("td")
    for p in th:
        temp_list.append(p.text)
    for p in td:
        temp_data.append(p.text)
    for p in th:
        scrape_list_test.append(p.text)
    scrape_list.append(temp_list)
    scrape_data.append(temp_data)
```

```python
scrape_list_series = pd.Series(scrape_list_test)
print(scrape_list_series.value_counts())
```

```python
print(scrape_list)
print(len(scrape_list))
```

```python
df = pd.Series(scrape_list)
print(df.head())
```
